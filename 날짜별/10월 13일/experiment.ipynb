```python
import pandas as pd
import tqdm
from bs4 import BeautifulSoup

page = open('plot_by_n.html', 'rt', encoding='utf-8').read()
soup = BeautifulSoup(page, 'html.parser')

df = pd.DataFrame([])
for i in range(2, 9):
  name = soup.select(f'body > div:nth-child(17) > table > tbody > tr > td:nth-child(1) > table > tbody > tr:nth-child(1) > th:nth-child({i})')[0].text.replace(' ', '').replace('\n', '')
  res = []
  for j in tqdm.trange(0, 317):
    if j % 41 == 40: continue
    res.append(soup.select(f'body > div:nth-child(17) > table > tbody > tr > td:nth-child(1) > table > tbody > tr:nth-child({j+2}) > td:nth-child({i}) > span:nth-child(2)')[0].text)
  df[name] = res

df
```

```python
df.to_csv('plot_by_n.csv')
```
