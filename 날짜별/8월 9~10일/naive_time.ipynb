```python
import matplotlib.pyplot as plt

plt.plot(M, avg_time, label="time")
plt.xlabel("M")
plt.ylabel("Avg. time")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
line_fitter = LinearRegression()
X = np.array([[M[i] * M[i]] for i in range(10)])
y = np.array(avg_time)
line_fitter.fit(X, y)
print(line_fitter.coef_, line_fitter.intercept_)
```
